```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the number of layers
num_layers = 3

# Create sample global map data for each layer
data = [np.random.rand(180, 360) for _ in range(num_layers)]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define the extent of the maps
lon = np.linspace(-180, 180, 360)
lat = np.linspace(-90, 90, 180)
lon, lat = np.meshgrid(lon, lat)

# Adjust vertical separation between each map
layer_heights = np.linspace(0, 100, num_layers)  # Change 100 to adjust the separation

for i in range(num_layers):
    X, Y = np.meshgrid(range(data[i].shape[1]), range(data[i].shape[0]))
    Z = np.ones_like(X) * layer_heights[i]
    
    # Plotting each layer with global map data
    ax.plot_surface(X, Y, Z, rstride=5, cstride=5, facecolors=plt.cm.viridis(data[i]), shade=False)

# Set viewing angle to better show the layers
ax.view_init(30, 120)

# Set labels
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Layer')

# Set the limits for the latitude and longitude labels
ax.set_xlim(0, data[0].shape[1])
ax.set_ylim(0, data[0].shape[0])
ax.set_zlim(0, 100)

plt.show()
```
